### Normalization

```python
x_trainSet = (x_trainSet - np.mean(x_trainSet, axis=0)) /np.std(x_trainSet, axis=0)
x_testSet = (x_testSet - np.mean(x_trainSet, axis=0)) /np.std(x_trainSet, axis=0)
```

Augment training and testing for fast matrix operation

```python
x_trainSet_aug = np.hstack((np.ones((x_trainSet.shape[0],1)), x_trainSet))
x_testSet_aug = np.hstack((np.ones((x_testSet.shape[0],1)), x_testSet))
```

## Logistic Regression Algorithm

### Sigmoid Function

```python
def h(x, weight):
    z = np.dot(x, weight.T)
    h = 1.0 / (1.0 + np.exp(-z))
    return h
```

### Cost Function

```python
def cost(x,y,weight):
    N = y.shape[0]
    #Logistic Likelihood
    loglike = 1.0/(2*N) * np.sum((h(x,weight)-y)**2)
    return loglike
```

### Gradient Descent

```python
def logisticRegression(x,y,weight,learn_rate,reg,maxIteration = 1000):
    y = y[:,np.newaxis]
    N = y.shape[0]
    i = 0
    while True:
        print('{:>2d} loss = {:>8f}'.format(i, cost(x,y,weight)))
        weight_prev = weight.copy()
        error = (h(x,weight)- y)
        weight = weight - (learn_rate*1.0)/N * np.sum(error * x, axis=0)
        i=i+1
        if (np.max(np.abs(weight - weight_prev)) < reg) or (i > maxIteration):
            break
    return weight
```

Learn parameters based off random parameter vector generated

```python
weight = np.random.randn(1, x_trainSet_aug.shape[1])
weight = logisticRegression(x_trainSet_aug, y_trainSet, weight,0.1,0.02)
```

Test classifier performance

```python
y_predict = (h(x_testSet_aug, weight) >= 0.5) [:,0]
test_error = np.sum(y_testSet != y_predict) / y_testSet.shape[0]
print('Logistic Regression test error = {:.4f}'.format(test_error))
```

### Confusion Matrix

```python
print(metrics.confusion_matrix(y_testSet, y_predict))
```
